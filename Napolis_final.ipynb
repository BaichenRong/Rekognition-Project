## Section 3.Visualization and Analysis - What are some common traits that influence the Beijing Housing prices?

In this case, we have chosen the labels architecture, housing, urban and mansion to see how the confidence intervals of these change as the with the different prices. These were selected because they were present in first 4 Houses above. For the ones that did not have this label, the value considered is the mean of the others.

```python
Architecture = {
    'Name': ["House1", "House2", "House3", "House4","House6","House8","House9"],
    'Price': [2500000,3021312,2300423,3120193,668000,955000,3012312],
    'ConfInterval' :[56.0834465,72.17907715,97.73561859,60.64037323,69.74581909179688,84.87091827392578,66.98990631103516]
}
Architecturedf = pd.DataFrame (Architecture, columns = ['Name','Price','ConfInterval'])

print (Architecturedf)

sns.lmplot(x='Price', y='ConfInterval', data=Architecturedf, ci=None)
```

Archtecture Confidence level variable showed a small negative relationship with the housing price.

```python
Housing = {
    'Name': ["House1", "House2", "House3", "House4","House5","House7","House9"],
    'Price': [2500000,3021312,2300423,3120193,2801231,742000,985000],
    'ConfInterval' :[83.19922892,83.14572906,67.2434082,99.2085495,81.99801635742188,64.9547119140625,94.335693359375]
}
Housingdf = pd.DataFrame (Housing, columns = ['Name','Price','ConfInterval'])

print (Housingdf)
sns.lmplot(x='Price', y='ConfInterval', data=Housingdf, ci=None)
```

Housing variable showed a relatively positive relationship with the housing price. It makes sense to us because housing confidence level may determined by the outlook and size of the house. Large house may have a higher price.

```python
Mansion = {
    'Name': ["House1", "House2", "House3", "House4"],
    'Price': [2500000,3021312,2300423,3120193],
    'ConfInterval' :[82.9217071533,82.31316376,67.2434082,99.2085495]
}
Mansiondf = pd.DataFrame (Mansion, columns = ['Name','Price','ConfInterval'])

print (Mansiondf)
sns.lmplot(x='Price', y='ConfInterval', data=Mansiondf, ci=None)
```

As seen, there are no labels which all 9 houses have in common. There are 2 (urban and Building) that 8/9 have in common; however, there are many labels which are not common to even 6 of the houses. As such, it might not be a good idea to see the trend by filling in the mean values for confidence intervals as this may skew the mean after all the values have been replaced. Another reason for this is that some labels are specific to houses such as waterfront. In this case, it would not make sense to have a mean confidence interval across the board as not every house is on the waterfront. An example of taking the mean value of the confidence interval for everyone and only using the images with the specified label is shown below. The label used here is downtown

```python
Downtown = {
    'Name': ["House1", "House2", "House3", "House4","House5","House6","House7","House8","House9"],
    'Price': [2500000,3021312,2300423,3120193,3333000,668000,742000,955000,985000],
    'ConfInterval' :[56.0752182,73.38232498,55.55601883,73.38232498,73.38232498,86.82359314,91.16691589,77.28987885,73.38232498]
}
Downtowndf = pd.DataFrame (Downtown, columns = ['Name','Price','ConfInterval'])

print (Downtowndf)
sns.lmplot(x='Price', y='ConfInterval', data=Downtowndf, ci=None)
```

As seen in the two graphs, there is a much stronger downwards trend between price and the confidence level of the downtown label when the images without this label are excluded as compared to when they are not.

Another interesting label is the city label. 7/9 images have this label; however, all of these images should have the city label as all the houses are in Beijing, which is a city. ### A .Downtown Label

By convension, Beijing citizens regard Dongcheng district as the representative city center. Thus, in our analysis, we will treat a remarkable place of interet in DongCheng District - Temple of heaven as the city center and measured each houses' distances to it.

```python
picture = fs.open('s3://beijinghousingprice/Beijing_district.png')
image = Image.open(picture)
display(image)
```

Picture of Temple of Heaven

By using Baidu map[link]https://map.baidu.com/ and addresses provided by Lianjia of each house, we can calculate the distance.  Thus, we generated the following barplot and dataframe.
Distance is in Kilometers

```python
Downtown1 = {
    'Name': ["House1", "House2", "House3", "House4","House5","House6","House7","House8","House9"],
    'Distance': [9.4,21.2,7.4,19.3,22.3,18.3,32.1,25.4,24.5],
    'ConfInterval' :[56.0752182,73.38232498,55.55601883,73.38232498,73.38232498,86.82359314,91.16691589,77.28987885,73.38232498]
}
```

```python
fig = plt.figure()
Name = ["House1", "House2", "House3", "House4","House5","House6","House7","House8","House9"]
Distance = [9.4,21.2,7.4,19.3,22.3,18.3,32.1,25.4,24.5]
ax = fig.add_axes([0,0,1,1])
ax.bar(Name,Distance)
ax.set_ylabel('Distance in Kilometers')
ax.set_title('Barplox of Beijing Houses by distance')
plt.show()
```

Let's apply a simple linear regression model to check the relationship between confidence level of downtown and distance

```python
from scipy.stats import linregress
xs1 = [56.0752182,73.38232498,55.55601883,73.38232498,73.38232498,86.82359314,91.16691589,77.28987885,73.38232498]
ys1 = [9.4,21.2,7.4,19.3,22.3,18.3,32.1,25.4,24.5]
res1 = linregress(xs1, ys1)
res1
```

```python
plt.plot(xs1, ys1, 'o', alpha=0.9)
fx = np.array([min(xs1), max(xs1)])
fy = res1.intercept + res1.slope * fx
plt.plot(fx, fy, '-')
plt.xlabel('Confidence level of downtown')
plt.ylabel('Distance in km')
plt.title('Linear regression: relationship between confidence level of downtown and distance in km');
```

From the bar plot and the linear regression model we set up, we can conclude that confidence level of downtown have a positive relationship with the distance level. It has a positive slope of 0.55. Thus, our hypothesis is correct! If the houses is more far away from the city center pushing distance value up, that means the level of downtown will go up. AWS successfully captured the degree of downtown label for our houses!

### B. Housing label

As we supposed, housing size should have a postive relationship with the confidence level. By collecting all the 9 houses data on Lianjia, we generated the following dataset.

```python
Housing = {
    'Name': ["House1", "House2", "House3", "House4","House5","House7","House9"],
    'Size_m_square': [231, 219, 221, 285, 196, 162, 303],
    'ConfInterval' :[83.19922892,83.14572906,67.2434082,99.2085495,81.99801635742188,64.9547119140625,94.335693359375]
}
```

```python
subset = pd.DataFrame(Housing, columns = ['Size_m_square','ConfInterval'])
subset
```

Having a look of the correlation. It has a number of 0.84 which is positively correlated.

```python
subset.corr()
```

```python
xs = subset['ConfInterval']
ys = subset['Size_m_square']
np.corrcoef(xs, ys)
```

```python
res1 = linregress(xs, ys)
res1
```

Generating a simple linear regression to check the degree of correlation

```python
plt.plot(xs, ys, 'o', alpha=0.9)
fx = np.array([min(xs), max(xs)])
fy = res1.intercept + res1.slope * fx
plt.plot(fx, fy, '-')
plt.xlabel('Confidence level of House')
plt.ylabel('Size in meter square')
plt.title('Linear regression: relationship between confidence level of House and Size in meter square');
```

Housing label did have a strong positive correlation between confidence level of House label and House size. It has a slope of 3.26. The regression model shows that when AWS detected a rising confidence level of house label, it means that size of corresponding house is larger!
